# Music Sales across Decades

- <a href="https://raw.githubusercontent.com/sandeepmj/datasets/main/1967_hits.csv">Most sold LPs</a> in 1967.
- <a href="https://raw.githubusercontent.com/sandeepmj/datasets/main/most-streamed-2023.csv">Most streamed albums</a> in 2023.

Which albums stood out (AS IN THE BIGGEST)in terms of sales?

```python
## import libraries and packages
import pandas as pd

## mad math functions package
from scipy.stats import zscore 


## run the display code here
pd.options.display.float_format = '{:,.2f}'.format
```

```python
#read data frame for 1967
lps = pd.read_csv("https://raw.githubusercontent.com/sandeepmj/datasets/main/1967_hits.csv")
lps.head(10)
```

```python
#read data frame for 2023
dfs = pd.read_csv("https://raw.githubusercontent.com/sandeepmj/datasets/main/most-streamed-2023.csv")
dfs.head(10)
```

```python
#find mean and std for 2023 albums
dfs[['streams_2023']].agg(["mean", "std"])
```

```python
#find mean and std for 1967 albums
lps[['sales']].agg(["mean", "std"])
```

```python
dfs["zscore"] = zscore(dfs["streams_2023"])
dfs
dfs.sort_values(by = ["zscore"], ascending = False).head(10)
```

```python
lps["zscore"] = zscore(lps["sales"])
lps
lps.sort_values(by = ["zscore"], ascending = False).head(10)
```

```python
df = pd.concat([dfs, lps], ignore_index = True)
df
df.sort_values(by = ["zscore"], ascending = False).head(10)
```
